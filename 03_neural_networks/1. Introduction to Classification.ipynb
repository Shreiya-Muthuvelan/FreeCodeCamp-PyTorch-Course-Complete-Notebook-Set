## What is Classification?

Classification refers to a type of problem that involves our ML algorithm trying to predict a category


### **Types of Classification**
- **Binary Classification**- There are two possible categories that can be output 

- **Multiclass Classification**- There are multiple categories that can be output. Example: Is this a sushi, pizza or steak? , the model assigns one cateogry/label to the picture 

- **Multilabel Classification**- Multiple label options per sample. Example: Deciding the tags for a article, your article can have multiple tags


### **Classification inputs and outputs**

Lets say we need to build a app that takes a picture of a food item and predicts what food it is, what we would do is:

         __________
        |          |
Input ->| ML Algo  | ->  Output 
        |__________|       

We need to convert the picture to a numeric form, 
1. We need take the picture and resize it to (224,224,3) - > (W,H,C)     
2. Numerically encode the pixel values 
3. Feed the encoded pixel values to the algorithm
4. The model outputs a probability number of which class of food the picture belongs to 
5. We can write some code to turn those probability values into actual categories and output the categories

### Architecture of a classification model

- Input layer shape (in_features): same as number of features . Remains same for both binary and multiclass classification
- Hidden layers(s)- Number of hidden layers is problem specific, minimum =1, maximum= unlimited
- Nuerons per hidden layer- Number of neurons per hidden layer is also problem specific . Can range from 10 to 512.
- Output layer shape(out_features): For binary classification it will be 1 (one class or other) , for multiclass classification it is 1 per class 
- Hidden layer activation : The activation function for hidden layer is normally ReLU
- Output activation : Output activation for binary classification is usually sigmoid. For multliclass classification it is softmax
- Loss function: For binary classification it is binary cross entropy and for multliclass classficiation it is cross entropy
- Optimizer : Common optimizers include SGD and Adam
