## Reproduciblity 

Trying to take the random out of random.

In short how a nueral network learns:
'start with random numbers->tensor operations->update random numbers to try to make them better representations of the data->repeat'

To reduce the randomness in NN, we have a concept called random seed



```python
import torch

random_tensor_A=torch.rand(3,4)
random_tensor_B=torch.rand(3,4)
print(random_tensor_A)
print(random_tensor_B)
```

```python
print(random_tensor_A==random_tensor_B)
```

```python
# reproducible tensors
random_seed=42
torch.manual_seed(random_seed) # works only for one block of code
random_tensor_C=torch.rand(3,4)

torch.manual_seed(random_seed) #have to assign seed again to get same tensor
random_tensor_D=torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
```

```python
print(random_tensor_C==random_tensor_D)
```
