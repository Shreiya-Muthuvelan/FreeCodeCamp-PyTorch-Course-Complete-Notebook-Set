## Computer Vision

Computer vision is a branch of artificial intelligence that helps computers interpret and understand visual data like images and videos

**Topics to be covered in this section:**

- Working with vision datasets using torchvision.datasets
- Architecture of a convolutional neural netwotk (CNN) with PyTorch
- End to end multi class image classification problem
- Steps in modelling with CNNs in PyTorch


### **Computer vision inputs and outputs**

- **Input**: This would typically be a tensor of the image we are representing that is its numerical encoding


- **Output**: Probability prediction of which class the image will belong to. We will use this predicted probability to find the class of the image 


### **Model**

The machine learning algorithm that usually predicts the probability of a class is a convolutional neural network (CNN)

### **Input and output shapes**

We want our input and output shapes to line up like previously. 

The workflow would remain the same, numerically encoder your data, feed it into a model, and then make sure the output shape fits your specific problem
